```python
#d.
d = np.random.normal(loc=-3,scale=math.sqrt(10),size=10)

print("Mean: ",np.mean(d))
print("Variance: ",np.var(d))

plt.hist(d)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Sample Size 10')
```

```python
#e.
d = np.random.normal(loc=-3,scale=math.sqrt(10),size=100)

print("Mean: ",np.mean(d))
print("Variance: ",np.var(d))

plt.hist(d)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Sample Size 100')
```
